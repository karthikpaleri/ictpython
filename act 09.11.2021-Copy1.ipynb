```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('diabetes_model.csv')
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
sns.countplot(df['Outcome'])
plt.show()
```

```python
ax = sns.countplot(x="Outcome",data=df)

for p in ax.patches:
    ax.annotate('{:.0f}'.format(p.get_height())),(p.get_x()+0.25,p.get_height())
    
plt.show()
```

```python
sns.boxplot(x = 'Outcome',y = 'Age', data = df)
plt.title('Age vs Outcome')
plt.show()
```

```python
##There is a chance of person being diabetic with increase in his age
```

```python
my_pal = {0:"lightgreen",1:"lightblue"}
sns.boxplot(x ='Outcome',y = 'DiabetesPedigreeFunction',data = df,palette = my_pal )
plt.title('DPF vs Outcome')
plt.show()
```

```python
sns.scatterplot(df ['Insulin'], df['Glucose'],hue = df['Outcome'])
plt.xlabel('Insulin')
plt.ylabel('Glucose')
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(df.corr(),annot=True, linewidth=0.2,fmt='.1f',cmap='coolwarm')
plt.show()
```
